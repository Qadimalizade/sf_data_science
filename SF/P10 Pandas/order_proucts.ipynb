```python
import pandas as pd
orders = pd.read_csv('data/orders and pr/orders.csv', sep=';')
display(orders)
```

```python
product_df = pd.read_csv('data/orders and pr/products.csv', sep=';')
display(product_df)
```

```python
orders_df = pd.concat(
    [orders, product_df],
    ignore_index=True
)
display(orders_df)
```

```python
orders_products = orders_df.merge(
    product_df, 
    left_on='ID товара',
    right_on='Product_ID',
    how='left')
orders_products.tail(1)['Order ID']
```
